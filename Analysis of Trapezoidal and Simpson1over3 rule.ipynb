```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
def Simpson(xi, xf, n, f):
    x = np.linspace(xi, xf, n)
    h = (xf - xi)/(n - 1)
    y = f(x)
 
    sum1 = 0
    for i in range(1, n-1, 2):
        sum1 += y[i]
 
    sum2 = 0
    for i in range(2, n-1, 2):
        sum2 += y[i]
     
    #Simpson 1/3 rule    
    sum = h*(y[0] + 4*sum1 + 2*sum2 + y[n-1])/3
    
    return sum
 
def Trapezoidal(xi, xf, n, f):
    x = np.linspace(xi, xf, n)
    h = (xf-xi)/(n-1)
    y = f(x)
    sum1 = 0
    for i in range(1, len(y)-1):
        sum1 += y[i]

    # Trapezoidal formula:
    sum2 = 0.5*h*(y[0]+2*sum1+y[len(y)-1])
    return sum2
```

```python
#Function defination
def f(x):
    return np.sin(x)
 
# Limit of integration
xi = 0
xf = np.pi
 
# List to plot 
N = []
SS = []
TT = []

for i in range(2, 20):
    S = Simpson(xi, xf, 2*i-1, f)
    T = Trapezoidal(xi, xf, i, f)
    N.append(i)
    SS.append(S)
    TT.append(T)

#Plot
plt.rcParams["figure.figsize"] = (12,4)
plt.plot(N, TT, label = "Trapezoidal")
plt.plot(N, SS, label = "Simpson")
plt.legend(loc = 'best')
plt.ylabel("Value of intregration")
plt.xlabel('No of points')
plt.show()
```

```python
#Function defination
def f(x):
    return np.exp(-x*x)/np.sqrt(np.pi)
# no of Points
n = 10
 
# List to plot 
N = []
SS = []
TT = []
a = 0.1
while ( a < 10):
    S = Simpson(-a, a, n, f)
    T = Trapezoidal(-a, a, n, f)
    N.append(a)
    SS.append(S)
    TT.append(T)
    a += 0.1

#Plot
plt.rcParams["figure.figsize"] = (12,8)
plt.plot(N, TT, label = "Trapezoidal")
plt.plot(N, SS, label = "Simpson")
plt.legend(loc = 'best')
plt.ylabel("Value of intregration")
plt.xlabel('Value of the limit')
plt.show()
```
