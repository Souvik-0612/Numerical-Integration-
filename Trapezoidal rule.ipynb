In the above picture I was able only to show 6 intervals. But it increses as we are going to decrese the size of the interval.
   Let $h$ be equally spaced interval. So $(x_0-x_1)=(x_1-x_2)=...=(x_{n-1}-x_n)=h$

   1.  **Approximation integral:**
   $$\int_{x_{n-1}}^{x_n} y dx \approx \frac {h}{2} (y_{n-1}+y_n)$$
  Using the above relation we can write:
    $$\int_{x_{0}}^{x_n} y dx \approx \frac {h}{2} [y_0+2(y_1+y_2+...+y_{n-1})+y_n]$$
    
    
   2.  **Error calculation** (From S S Sastry "Introductory Methods to Numerical Analysis")
    $$\int_{x_0}^{x_1} y dx = \int_{x_{0}}^{x_1}[y_0+(x-x_0)y_0'+\frac {(x-x_0)^2}{2!}y_0''+...]dx= hy_0+h^2\frac{y_0'}{2!}+h^3\frac{y_0''}{3!}+...$$
    
  But our approximation for this interval was $\frac{h}{2}(y_0+y_1)$. Using Taylor seires
  
   $$\frac{h}{2}(y_0+y_1)=\frac{h}{2}(y_0+y_0+hy_0'+\frac{h^2}{2!}y_0''+\frac{h^3}{3!}y_0'''+...)$$
   Only consider upto 2nd order derivatives
    $$E_1 = -\frac{1}{12}h^3 y_0''$$
    $$E_2 =-\frac{1}{12}h^3 y_1''$$
    $$E_3 =-\frac{1}{12}h^3 y_2''$$
    $$E_4 =-\frac{1}{12}h^3 y_3''$$
    $$.......$$
    $$E_n =-\frac{1}{12}h^3 y_{n-1}''$$
   Therefore **total error** is 
    $$E = E_1+E_2+E_3+...+E_n$$
    $$E = -\frac{1}{12}h^3[y_0''+y_1''+y_2''+...+y_{n-1}'']$$
